```python
try:
    logo=np.load('numpy-logo.npy')

    #Display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)

    dark_logo = 1-logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Logo")
    plt.grid(False)


except FileNotFoundError:
    print("numpy logo file not found")

```
